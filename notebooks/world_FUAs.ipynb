```python
import pandas as pd
from opencage.geocoder import OpenCageGeocode
import time

# Your OpenCage API key
api_key = '1f565f92418545f6ac65946f084e25b9'  # Replace with your actual OpenCage API key
geocoder = OpenCageGeocode(api_key)

# Assuming unique_FUAs is your DataFrame and has 'FUA' (City) and 'Country' columns
def get_coordinates(city_name, country_name):
    # Combine city and country for more accurate geocoding
    location_query = f"{city_name}, {country_name}"
    result = geocoder.geocode(location_query)
    if result:
        # Extract latitude and longitude from the result
        latitude = result[0]['geometry']['lat']
        longitude = result[0]['geometry']['lng']
        return latitude, longitude
    else:
        return None, None

# Applying the function to get coordinates for each city-country pair
unique_FUAs[['Latitude', 'Longitude']] = unique_FUAs.apply(
    lambda row: pd.Series(get_coordinates(row['FUA'], row['COUNTRY'])),
    axis=1
)

# Display the updated DataFrame
unique_FUAs
```
