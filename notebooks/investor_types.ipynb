```python
import pandas as pd
import numpy as np

#data cleaning
df = pd.read_csv('https://raw.githubusercontent.com/veghmark12/r_d_investments/refs/heads/main/df_with_closest_fua_extended_2023.csv')
```

```python
df['EACH INVESTOR TYPES'] = df['EACH INVESTOR TYPES'].replace('angel_fund', 'angel', regex=True)
df['EACH INVESTOR TYPES'] = df['EACH INVESTOR TYPES'].replace('corporate_venture_fund', 'venture_capital', regex=True)
df['EACH INVESTOR TYPES'] = df['EACH INVESTOR TYPES'].replace('sovereign_wealth_fund', 'venture_capital', regex=True)
df['EACH INVESTOR TYPES'] = df['EACH INVESTOR TYPES'].replace('incubator', 'accelerator&incubator', regex=True)
df['EACH INVESTOR TYPES'] = df['EACH INVESTOR TYPES'].replace('accelerator', 'accelerator&incubator', regex=True)
df['EACH INVESTOR TYPES'] = df['EACH INVESTOR TYPES'].replace('accelerator&incubator&incubator', 'accelerator&incubator', regex=True)
df['EACH INVESTOR TYPES'] = df['EACH INVESTOR TYPES'].replace(['venture_capital','angel','private_equity','family_office','corporate','investment fund'], 'investment entities', regex=True)
df['EACH INVESTOR TYPES'] = df['EACH INVESTOR TYPES'].replace(['accelerator&incubator','workspace','workspace location'], 'support programs', regex=True)
df['EACH INVESTOR TYPES'] = df['EACH INVESTOR TYPES'].replace('support programs location', 'support programs', regex=True)
df['EACH INVESTOR TYPES'] = df['EACH INVESTOR TYPES'].replace(['government','crowdfunding','non-profit'], 'public and non-profit', regex=True)
df['EACH INVESTOR TYPES'] = df['EACH INVESTOR TYPES'].replace('public and non-profi', 'public and non-profit', regex=True)
df['EACH INVESTOR TYPES'] = df['EACH INVESTOR TYPES'].replace('public and public and non-profitt', 'public and non-profit', regex=True)
df['EACH INVESTOR TYPES'] = df['EACH INVESTOR TYPES'].replace('public and non-profitt', 'public and non-profit', regex=True)
df['EACH INVESTOR TYPES'] = df['EACH INVESTOR TYPES'].replace('public and non-profit & public and non-profit', 'public and non-profit', regex=True)
df['EACH INVESTOR TYPES'] = df['EACH INVESTOR TYPES'].replace(['advisor','service provider'], 'advisory and service provider', regex=True)
df['EACH INVESTOR TYPES'] = df['EACH INVESTOR TYPES'].replace('advisory and advisory and service provider', 'advisory and service provider', regex=True)
df['EACH INVESTOR TYPES'] = df['EACH INVESTOR TYPES'].replace('company', 'investment entities', regex=True)
df['EACH INVESTOR TYPES'] = df['EACH INVESTOR TYPES'].replace('university', 'public and non-profit', regex=True)
```

```python
# Extract the values from the column into a list
values_list = df['EACH INVESTOR TYPES'].tolist()
```

```python
# Ensure all entries are strings; replace None or NaN with an empty string
cleaned_list = [str(item) if item is not None else '' for item in values_list]

# Split the values by both semicolons and commas, and flatten the result
separated_list = [item for sublist in cleaned_list for item in sublist.replace(';', ',').split(',')]

# Get unique values
real_unique_values = pd.Series(separated_list).unique()

# Output unique values
print(real_unique_values)

df
```

```python
import pandas as pd

# Ensure all values in the column are strings and replace NaN with an empty string
df['EACH INVESTOR TYPES'] = df['EACH INVESTOR TYPES'].fillna('').astype(str)

# Categories to count
categories = ['public and non-profit', 'investment entities', 'support programs','advisory and service provider']

# Count occurrences of each category
for category in categories:
    df[category] = df['EACH INVESTOR TYPES'].apply(lambda x: x.split(';').count(category))

df
```
